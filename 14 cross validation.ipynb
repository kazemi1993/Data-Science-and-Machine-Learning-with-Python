```python
import pandas as pd

file_path = '/content/drive/MyDrive/Colab Notebooks/ml/AMES_Final_DF.csv'

df = pd.read_csv(file_path)

df.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df.info()
```

```python
X = df.drop('SalePrice',axis=1)
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score


# Create a linear regression model
lm = LinearRegression()
```

```python
# # Perform cross-validation
# cv_scores = cross_val_score(lm, scaled_X_train, y_train, cv=5) # 5-fold cross-validation

# # # Print the cross-validation scores
# print(cv_scores)
# print(f'Average CV Score: {np.mean(cv_scores)}')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# Fit the model to the training data
lm.fit(scaled_X_train, y_train)

# Make predictions on the test set
y_pred = lm.predict(scaled_X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)


print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared (R2): {r2}')
```
