Filling in Data or Dropping Data?

Dropping Data

```python
# Print the sum of null values for each column
print(df.isnull().sum())

# Drop columns with more than 5% missing values
df = df.dropna(thresh=len(df)*0.95, axis=1)

# Drop rows with any missing values
df = df.dropna()

# Print the sum of null values for each column again to verify
print(df.isnull().sum())
```

Filling in Data

```python
# prompt: # Fill missing values with the mean for numerical columns
# show me before and after missing data

import numpy as np
# Print the sum of null values for each column before filling
print("Missing values before filling:\n", df.isnull().sum())

# Fill missing values with the mean for numerical columns
for col in df.select_dtypes(include=np.number):
    df[col] = df[col].fillna(df[col].mean())

# Print the sum of null values for each column after filling
print("\nMissing values after filling:\n", df.isnull().sum())
```
