```python
import pandas as pd

file_path = '/content/drive/MyDrive/Colab Notebooks/ml/Ames_NO_Missing_Data.csv'


df = pd.read_csv(file_path)

df.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_nums = df.select_dtypes(exclude='object')
df_nums
```

```python
df_objs = df.select_dtypes(include='object')
df_objs
```

```python
df_objs = pd.get_dummies(df_objs)
df_objs
```

```python
final_df = pd.concat([df_nums,df_objs],axis=1)
final_df
```

```python

from sklearn.preprocessing import OrdinalEncoder

# Create an instance of OrdinalEncoder
ordinal_encoder = OrdinalEncoder()

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Fit and transform the categorical columns
df[categorical_cols] = ordinal_encoder.fit_transform(df[categorical_cols])
```

```python

from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Iterate through categorical columns and apply label encoding
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])
```

```python

for col in df.columns:
    if df[col].dtype == 'object':
        mean_encoding = df.groupby(col)['SalePrice'].mean()
        df[col] = df[col].map(mean_encoding)
```
