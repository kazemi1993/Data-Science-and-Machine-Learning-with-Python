### " At this point, only a few countries have missing values. You can either remove these countries or fill in the gaps using any method you prefer. To keep things simple, we'll just drop them."



```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
X = df.drop("Country",axis=1)
X
```

```python
X = pd.get_dummies(X)
```

```python
X.head()
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
scaled_X
```

### Creating and Fitting Kmeans Model




```python
from sklearn.cluster import KMeans

#`ssd` stands for **Sum of Squared Distances**.
# It's a list that stores the sum of squared distances of samples to their closest cluster center for different values of *k* (the number of clusters) in the KMeans clustering algorithm.
ssd = []

for k in range(2,30):

    model = KMeans(n_clusters=k)


    model.fit(scaled_X)

    # model.inertia_ represents the sum of squared distances of samples to their closest cluster center.
    # Lower inertia generally indicates better-defined clusters, where data points within a cluster are closer to the cluster's centroid.
    # This metric is used in the elbow method to determine the optimal number of clusters (k) for KMeans clustering.

    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,30),ssd,'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances")
```

```python
pd.Series(ssd).diff().plot(kind='bar')
```
