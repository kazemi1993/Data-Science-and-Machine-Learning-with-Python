```python
import pandas as pd

file_path = '/content/drive/MyDrive/Colab Notebooks/ml/cancer-tumor-data-features.csv'


df = pd.read_csv(file_path)

df.head()
```

### Scaling Data



```python
# Remove non-numeric columns if any
df = df.select_dtypes(include=np.number)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(df)
scaled_X
```

## PCA with Scikit-Learn

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_X)


plt.figure(figsize=(4,3))
plt.scatter(principal_components[:,0],principal_components[:,1])
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
total_variance_explained = np.sum(pca.explained_variance_ratio_)

print(f"Explained variance ratio of each principal component: \n {explained_variance_ratio}")
print("\n")
print(f"Total variance explained by the two principal components: {total_variance_explained}")
```

```python
explained_variance = []
for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    explained_variance.append(np.sum(pca.explained_variance_ratio_))

plt.plot(range(1,30),explained_variance)
plt.xlabel("Number of Components")
plt.ylabel("Variance Explained");
```
